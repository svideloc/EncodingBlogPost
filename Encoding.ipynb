Target Encoding

```python
#import libraries
import pandas as pd
import numpy as np
from category_encoders import TargetEncoder

#creating dataframe with categoriccal features and target
Animal = ['cat', 'hamster', 'cat', 'cat', 'dog', 'hamster', 'cat', 'dog', 'cat', 'dog']
target =  [1, 0, 0, 1, 1, 1, 0, 1, 0, 0]
data = list(zip(Animal, target))
df = pd.DataFrame(data,
                columns=['Animal', 'Target'])

#encode
encoder = TargetEncoder()
df['Animal Encoded'] = encoder.fit_transform(df['Animal'], df['Target'])

df.head(10)
```

One Hot Encoding

```python
#creating dataframe with categoriccal features and target
import pandas as pd
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import LabelEncoder 

#Creating df
Animal = ['cat', 'hamster', 'cat', 'cat', 'dog', 'hamster', 'cat', 'dog', 'cat', 'dog']
target =  [1, 0, 0, 1, 1, 1, 0, 1, 0, 0]
data = list(zip(Animal, target))
df = pd.DataFrame(data,
                columns=['Animal', 'Target'])

#label encoding
le = LabelEncoder()
df['Animal Encoded'] = le.fit_transform(df.Animal)

#onehotencoding
encoder = OneHotEncoder(categories = 'auto')
X = encoder.fit_transform(
    df['Animal Encoded'].values.reshape(-1,1)).toarray()
dfonehot = pd.DataFrame(X)
df = pd.concat([df, dfonehot], axis =1)
df.columns = ['Animal','Target','Animal Encoded',
                     'isCat','isDog','isHamster']

df.head(10)
```
